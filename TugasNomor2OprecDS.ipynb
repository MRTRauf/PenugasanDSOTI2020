```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv(r"C:\Users\Lenovo\Downloads\Penugasan Divisi Data Science\Train.csv")
train.head()
```

```python
plt.scatter(train['nilai'],train['hasil'])
```

```python
train_copy = train.copy()
train_copy = train_copy.drop(columns = ['nama','kelas'])
train_copy
```

```python
x = train_copy.iloc[:, 0:1].values
y = train_copy.iloc[:, 1].values
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x, y)
```

```python
test = pd.read_csv(r"C:\Users\Lenovo\Downloads\Penugasan Divisi Data Science\Test.csv")
test
```

```python
test_copy = test.copy()
test_copy
```

```python
test_copy = test_copy.drop(columns = ['nama','kelas', 'hasil'])
test_copy
```

```python
pred = model.predict(test_copy)
pred = pd.DataFrame(pred, columns=['hasil'])
pred
```

```python
result = test
result
```

```python
result = test.drop('hasil', axis=1)
result
```

```python
result = pd.concat([result, pred], axis=1)
result
```
